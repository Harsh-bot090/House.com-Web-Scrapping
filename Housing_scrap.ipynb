```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

```python
# url = "https://housing.com/in/buy/searches/AB1M1Pskwz0ocdh7q42r5"

with open('mumbai.html','r',encoding='utf-8') as f:
    html = f.read()
# res = requests.get("https://housing.com/in/buy/searches/AB1M1Pskwz0ocdh7q42r5")
soup = BeautifulSoup(html , "lxml")

base_url = 'https://housing.com'
relative_url = soup.find_all('a' , '_j31f9d _c8dlk8 _g3l52n _csbfng _frwh2y T_e4485809 _ks15vq _vv1q9c _sq1l2s T_091c165f')
flats = pd.DataFrame()


count = 0

for i in relative_url:
    url = base_url + i.get('href')
    res = requests.get(url)
    soup = BeautifulSoup(res.text , "lxml")

    heading = soup.find('div' , 'css-1hidc9c')
    pata = soup.find('div' , 'css-1ty5xzi')
    points = soup.find('span' , '_csbfng _c8dlk8 _7l1qmd _h0yh40 _1q733bzg _4okucf5i T_f3f6bf63')
    try:
        name = heading.text 
    except:
        name = ' '
    try:
        address = pata.text
    except:
        address = ' '
    try:
        ratings = points.text
    except:
        ratings = ' '


    #  Property Overview
    desc_heading = soup.find_all('tr', 'T_2d90c4af _1voo1crd _fq0l1tcg _ao4416nl _1vgg1e5h _44yo10p5 _1gkhyfp0 T_2cb3d215 _7s5w8wlv _vhjp4b _gqexct _vysqn1 data-point')
    desc  = soup.find_all('tr' , 'T_c03272d6 _h31f4h _vybq9w data-point')
    count = count + 1      
    print(count)

    for i in desc:
            
        th = i.find('th' , '_7l11ef T_8cd85f88')
        td = i.find('div' , '_h3i2wt T_46c25558')
        try:
            if th.text == "Brokerage":
                brokerage =td.text
        except:
            brokerage = ' '
        try:
            if th.text == "Price":
                price = td.text
        except:
            price = ' '
        try:  
            if th.text == "Balcony":
                balcony = td.text
        except:
            balcony = ' '
        try:
            if th.text =="Bedrooms":
                bedrooms = td.text 
        except:
            bedrooms = ' '
        try:
            if th.text == "Bathrooms":
                bathrooms = td.text        
        except:
            bathrooms = ' '
            
    # Basic Info
    basic_info = soup.find_all('div', 'css-c2zxhw')
    for i in basic_info:
        bui_a = i.find('div', 'css-0')
        bui_a_value = i.find('div', 'css-1k19e3')

        try:
            if bui_a.text == "Built Up Area":
                built_up_area = bui_a_value.text
        except:
            built_up_area = " "
        
        try:
            if bui_a.text == 'Avg. Price':
                price_per_sqft = bui_a_value.text
        except:
            price_per_sqft = " "
        
        try:
            if bui_a.text == "Age of property":
                age_of_property = bui_a_value.text
        except:
            age_of_property = " "
        
        try:
            if bui_a.text == "Facing":
                facing = bui_a_value.text
        except:
            facing = " "
        
        try:
            if bui_a.text == "Furnishing":
                furnishing = bui_a_value.text
        except:
            furnishing = " "


    locals = soup.find_all('div' , 'T_091c165f _sq1l2s _vv1q9c _ks15vq T_d8632a7f _h3yh40 _7l1ulh _g3l52n _c8dlk8')
    try:
        locality  = [i.text.strip() for i in basic_info]
        
    except:
        locality = ' '

    features = soup.find_all('div' ,'T_5990d116 _1q731trj _6w1e54 _9scj1k _fycs5v _ks15vq _c81fwx _h3ftgi label')
    
    try:
        amenities= [i.text.strip() for i in features]
    except:
        amenites = ' '

    data = {
        "name":name,
        "address":address,
        "price_per_sqft":price_per_sqft,
        "brokerage":brokerage,
        "price":price,
        "balcony":balcony,
        "bedrooms":bedrooms,
        "bathrooms":bathrooms,
        "built_up_area":built_up_area,
        "age_of_property":age_of_property,
        "facing":facing,
        "furnishing":furnishing,
        "ratings":ratings,
        'locality':locality,
        'amenities':amenities
    }
    temp_df = pd.DataFrame.from_records([data])
    
    flats = pd.concat([flats, temp_df], ignore_index=True)
        
```
